```python
# 1. Import thư viện
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import LinearSVC
from sklearn.metrics import classification_report
import joblib
```

```python
# 2. Đọc dữ liệu
df = pd.read_csv("train.csv")
df.head(5)
```

```python
df.info()
```

```python
outcome_counts = df['label'].value_counts()

print("0:", outcome_counts.get(0, 0))
print("1:", outcome_counts.get(1, 0))
```

```python
# Step 1: Check for missing values
print("Missing values in each column:")
print(df.isnull().sum())
```

```python
# 3. Tiền xử lý (nếu cần)
df = df.dropna() 

# Nếu dùng tập train.csv
df['text'] = df['text'].astype(str)
```

```python
# 4. Vector hóa văn bản
vectorizer = TfidfVectorizer(max_features=5000)

# Nếu dùng tập public_train.csv
X = vectorizer.fit_transform(df['text'])
y = df['label']
```

```python
# 5. Chia dữ liệu train/test
X_train, X_test, y_train, y_test = train_test_split(X, y,train_size=0.8,  test_size=0.2, random_state=0)
```

```python
# 6. Huấn luyện mô hình SVM
svm_model = LinearSVC()
svm_model.fit(X_train, y_train)
```

```python
# 7. Đánh giá mô hình
y_pred = svm_model.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
# 8. Lưu mô hình và vectorizer
joblib.dump(svm_model, "toxic_model.joblib")
joblib.dump(vectorizer, "tfidf_vectorizer.joblib")
```
